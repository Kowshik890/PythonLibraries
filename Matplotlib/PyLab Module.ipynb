PyLab is a convenience module that bulk imports matplotlib.pyplot (for plotting) and NumPy (for Mathematics and working with arrays) in a single name space.

```python
from numpy import *
from pylab import *
x = linspace(-3, 3, 30)
y = x**2
plot(x, y)
show()
```

To plot symbols rather than lines, provide an additional string argument.

symbols -> - , â€“, -., , . , , , o , ^ , v , < , > , s , + , x , D , d , 1 , 2 , 3 , 4 , h , H , p , | , _

colors -> b, g, r, c, m, y, k, w

```python
from pylab import *
x = linspace(-3, 3, 30)
y = x**2
plot(x, y, 'r.') # r is for red color and dot (.) is for the curve
show()
```

```python
# Plots can be overlaid. Just use the multiple plot commands. 

from pylab import *
plot(x, sin(x))          # The color of the curse is blue by default 
plot(x, cos(x), 'r+')    # r for red and (+) is for the curve
plot(x, -sin(x), 'g*')  # g for green and (*) is for the curve
show()
```
