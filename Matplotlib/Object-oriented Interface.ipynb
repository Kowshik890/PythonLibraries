```python
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()     # To begin with, we create a figure instance which provides an empty canvas.

# The following table shows the additional parameters of "figure()"âˆ’

# Figsize    --->  (width,height) tuple in inches
# Dpi        --->  Dots per inches
# Facecolor  --->  Figure patch facecolor
# Edgecolor  --->  Figure patch edge color
# Linewidth  --->  Edge line width

ax = fig.add_axes([0, 0, 1, 1])     

# The add_axes() method requires a list object of 4 elements corresponding to left, bottom, width and height of the figure. 
# Each number must be between 0 and 1.

ax.set_title("Sine Wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
ax.plot(x, y)
```
