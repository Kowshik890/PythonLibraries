ax.bar(x, height, width, bottom, align)

x      = input value
height = output value (y axis)
width  = bar size in wide
bottom = from where it will start
align  = {'center', 'edge'}, optional, default 'center'

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['C', 'C++', 'C#', 'PHP', 'Java', 'Python']
students = [28, 17, 35, 42, 19, 25]
ax.bar(languages, students)
plt.show()
```

We can plot multiple bar charts with the thickness and the positions of the bars. 
For example:
The data variable contains three series of four values. The following script will show three bar charts of four bars. The bar will have a thickness of 0.25 units and each bar chart will be shifted 0.25 units from the previous one.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [[30, 25, 50, 20],        
        [40, 23, 51, 17],
        [35, 22, 45, 19]]
x = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.bar(x + 0.00, data[0], color = 'b', width = 0.25)  # width = 0.25 means shift the bar chart 0.25 units from the previous one.
ax.bar(x + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(x + 0.50, data[2], color = 'r', width = 0.25)

ax.set_xticks(x)
ax.set_xticklabels(['2015', '2016', '2017', '2018'])
fig.legend(labels = ('CS', 'IT', 'E&TC'), loc = 'upper right')
```

pyplot.bar() function allows to specify a starting value for a bar.

```python
import matplotlib.pyplot as plt
import numpy as np

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (35, 32, 34, 20, 25)
ind = np.arange(N)
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(ind, menMeans, width, color = 'r')
ax.bar(ind, womenMeans, width, bottom = menMeans, color = 'b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
#ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels = ['Men', 'Women'])
plt.show()
```
