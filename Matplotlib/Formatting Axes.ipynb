Sometimes, one or a few points are much larger than the bulk of data. For that, the scale of an axis needs to be set as logarithmic rather than the normal scale.

In Matplotlib, it is possible by setting 'set_xscale('log')' or 'set_yscale('log')' property of axes object to 'log'

For example, axes[1].set_yscale("log")   --> this syntax will help to set the large curve in y axis

To show some additional distance between axis numbers and axis label, 'labelpad' is used for (x or y or both axis)

For example, axes[0].xaxis.labelpad = 10   --> in x axis to maintain a distance between axis numbers and axis label.


```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize = (10, 4))
x = np.arange(1,5)

axes[0].plot(x, np.exp(x))
axes[0].plot(x, np.square(x))
axes[0].set_title('Normal Scale')

axes[1].plot(x, np.exp(x))
axes[1].plot(x, np.square(x))
axes[1].set_yscale("log")         # this syntax will help to set the large curve in y axis
axes[1].set_title('Logarithmic Scale')

axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].xaxis.labelpad = 20       # Distance between axis numbers and axis label is 20 in X axis from axes[0]

axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')

plt.show()
```

--> Axis spines are the lines connecting axis tick marks demarcating boundaries of plot area. 
--> The axes object has spines located at top, bottom, left and right.
--> Each spine can be formatted by specifying color and width.
--> Any edge can be made invisible if its color is set to 'None'

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('green')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color(None)    # This edge will be invisible
ax.spines['top'].set_color(None)      # This edge will be invisible
ax.plot([1, 2, 3, 4, 5])
plt.show()
```
