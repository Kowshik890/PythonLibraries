To set limit in Matplotlib, we can use "set_xlim(a, b)" and "set_ylim(a, b)" functions for both X and Y axis.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize = (12, 4))
x = np.arange(1, 10)

axes[0].plot(x, np.exp(x))
axes[0].set_title('Exp without limit')

axes[1].plot(x, np.exp(x))
axes[1].set_title('Exp')
axes[1].set_xlim(0, 10)
axes[1].set_ylim(0, 10000)

fig.tight_layout()
plt.show()
```
