 Twin axes is useful to have dual x or y axes in a figure. Matplotlib supports this with the 'twinx' and 'twiny' functions.

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
x = np.arange(1, 11)

a1 = fig.add_axes([0, 0, 1, 1])
a1.plot(x, np.exp(x))
a1.set_ylabel('Exp')

a2 = a1.twinx()     # the value of X axis will be remain same for both curves
a2.plot(x, np.log(x), 'ro-')
a2.set_ylabel('Log')

fig.legend(labels = ('exp', 'log'), loc = 'upper left')
plt.show()
```
