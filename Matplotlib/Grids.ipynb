The grid() function of axes object sets visibility of grids inside the figure to on or off. Also display major or minor or both ticks of the grid. Additionally color, linestyle and linewide propertities can be set in the grid() function.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize = (14, 4))   # figsize(width, height) --> figure size
x = np.arange(1, 11)

axes[0].plot(x, np.power(x, 3), 'g', lw = 2)        # lw = linewide (Width or density of the curved line))
axes[0].grid(True)                                  # grid(True) = default grid
axes[0].set_title('Default grid')

axes[1].plot(x, np.exp(x), 'r', lw = 3)             # this 'lw' is for the curved line
axes[1].grid(color = 'b', ls = '-.', lw = 1)        # for customize of grid() [color, ls(linestyle), lw are the parameters]
axes[1].set_title('Custom grid')

axes[2].plot(x, x, 'k', lw = 4)
axes[2].set_title('No grid')

plt.tight_layout()
plt.show()
```
